## Data Wrangling for individual City data set

### Steps taken for Data Wrangling

- Reading each individual city data set from AirBNB
- Removal of all unwanted columns from the data set
- Removal of all unwanted fields from each columns
- Converting category type column to Integer for columns like price, bathroom
- Taking care of all missing values in each column
- Renaming of couple of fields
- Merging of Listing data set as well as Calendar data set
- Deriving the new feature i.e Income

## Merging City Listing and Calendar Data set

```python
#City Listing Data set
city_master_data.head()
```

```python
#City Calendar Data set
city_master_calendar_data.head()
```

```python
#Merging the listing data set and calendar dara set and also summing up the price for respective ids
city_listing_calendar_temp = pd.concat([city_master_data, city_master_calendar_data]).groupby('id')\
                            ['income'].sum().reset_index()
```

```python
#Then LEFT merging with listing data set to contain only the ids present in listing data set
city_listing_calendar_data = pd.merge(city_master_data, city_listing_calendar_temp, how='left', on=['id'])
```
